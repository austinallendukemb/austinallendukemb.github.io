### Part 6

(3 points) Provide a query showing the invoices of customers who are from Germany. The resultant table should show the customer’s full name, Invoice ID, Date of the invoice and billing country.



```python
%%sql
-- Show the customer name, invoice id, invoice date and billing country for
-- customers from Germany
SELECT c.FirstName, c.LastName, i.InvoiceID, i.InvoiceDate, i.BillingCountry
FROM Customer AS c
INNER JOIN Invoice AS i using(CustomerID)
WHERE c.Country = 'Germany';
```

### Part 13


(2 points) Provide a query that shows total sales made by each sales agent.



```python
%%sql
-- Show the sales made by each agent
SELECT e.EmployeeId, e.FirstName, e.LastName, ROUND(COALESCE(SUM(i.Total), 0))
AS TotalSales
FROM Employee AS e
LEFT OUTER JOIN Customer AS c ON e.EmployeeId = c.SupportRepId
LEFT OUTER JOIN Invoice AS i USING(CustomerID)
GROUP BY e.EmployeeId
ORDER BY TotalSales DESC;
```

### Part 14


(3 points) Which sales agent made the most in sales in 2021? (Hint: count the total sales in 2021 grouped by sales agent.)




```python
%%sql
-- Determine which sales agent made the most sales in 2021
SELECT ID, Firstname, LastName, TotalSales FROM (
  SELECT e.EmployeeId AS ID, e.FirstName, e.LastName,
  ROUND(COALESCE(SUM(i.Total), 0)) AS TotalSales
  FROM Employee AS e
  LEFT OUTER JOIN Customer AS c ON e.EmployeeId = c.SupportRepId
  LEFT OUTER JOIN Invoice AS i USING(CustomerId)
  WHERE i.InvoiceDate BETWEEN datetime('2021-01-01 00:00:00')
  AND datetime('2021-12-31 00:00:00')
  GROUP BY e.EmployeeId
) ORDER BY TotalSales DESC LIMIT 1;
```

### Part 15

(3 points) Provide a query that shows the total sales per country. Which country’s customers spent the most?




```python
%%sql
-- Determine which country's customers spent the most
SELECT BillingCountry, ROUND(COALESCE(SUM(Total), 0)) AS TotalSales
FROM Invoice GROUP BY BillingCountry ORDER BY TotalSales DESC LIMIT 1;
```

Complete the following tasks:

### Part 1

(6 points) Show a table of ALL dogs and the treats with calories that they ate with column names dog, treat, calorie. A dog that did not eat any treats should still be present in the table.

```python
%sql sqlite:////content/drive/MyDrive/ColabNotebooks/homework_projects/pets.db
```

```python
%%sql
-- Find out the treats and calories each dog has consumed
SELECT d.name AS dog, t.name AS treat, COALESCE(t.calories, 0) AS calories
FROM dog AS d
LEFT OUTER JOIN dog_treat USING(dog_id)
LEFT OUTER JOIN treat AS t USING(treat_id)
ORDER BY dog, calories DESC;
```

### Part 2

(6 points) Show a table with two columns dog and total_calories where only dogs that have eaten more than 500 calories are displayed.

```python
%%sql
SELECT d.name AS dog, SUM(t.calories) AS total_calories
FROM dog AS d
LEFT OUTER JOIN dog_treat USING(dog_id)
LEFT OUTER JOIN treat AS t USING(treat_id)
GROUP BY d.dog_id HAVING total_calories > 500
ORDER BY total_calories DESC;
```
